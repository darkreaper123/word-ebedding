```python
#!pip install nltk
#!pip install PyDictionary
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
import string
from nltk.corpus import stopwords
#nltk.download('stopwords')
stop_words = stopwords.words("english")
print("Thêm vào các thư viện cần dùng")
#from google.colab import drive
#drive.mount('/content/drive')
irregular_verbs = pd.read_csv("drive/MyDrive/most-common-verbs-english.csv").values
print(irregular_verbs)
print(stop_words)
```

```python
#print(data)
#print('i' in stop_words)
data = pd.read_csv("drive/MyDrive/quotes_dataset.csv")["quote"].values
data = data[:1000]
#print(data)
alphabets = string.ascii_lowercase + string.ascii_uppercase
for i in range(len(stop_words)):
  j = 0
  while j < len(stop_words[i]):
    if stop_words[i][j] not in alphabets:
      stop_words[i] = stop_words[i][:j] + stop_words[i][j + 1:]
    else:
      j += 1
print(stop_words)
def preProcess(text):
    new_text = text.lower()
    #print(new_text)
    i = 0
    #print(alphabets)
    while i < len(new_text):
        if new_text[i] not in alphabets and new_text[i] != ' ':
            new_text = new_text[:i] + new_text[i + 1:]
            #i -= 1
        else:
            i += 1
    list_str = new_text.split(" ")
    new_text2 = ""
    for i in range(len(list_str)):
        for j in range(irregular_verbs.shape[0]):
            for k in range(1, irregular_verbs.shape[1]):
                if irregular_verbs[j, k] != irregular_verbs[j, k]:
                    continue
                if list_str[i] in irregular_verbs[j, k].split(" "):
                    #print(list_str[i])
                    list_str[i] = irregular_verbs[j, 0]
        if list_str[i] not in stop_words:
          #print(list_str[i])
          new_text2 += list_str[i] + " "
    new_text2 = new_text2[:-1]
    return new_text2
for i in range(len(data)):
     data[i] = preProcess(data[i])
#print(data[0].split(" ")[8])
#print(data[0].split(" ")[8]  == 'am')
#print(stop_words)
#print(data)
X_train = []
Y_train = []
word_index = {}
index_word = {}
V = 0
for line in data:
  for word in line.split(" "):
    if word not in word_index.keys():
       word_index[word] = V
       index_word[V] = word
       V += 1
def to_numeric(V, word, word_index):
  one_hot = np.zeros((V))
  one_hot[word_index[word]] = 1.
  return one_hot
window_size = 6
for line in data:
  line_list = line.split(" ")
  for i in range(len(line_list)):
    output = []
    for j in range(max(i - 3, 0), min(i + 3, len(line_list))):
        if line_list[i] != line_list[j]:
          X_train.append(to_numeric(V, line_list[i], word_index))
          Y_train.append(to_numeric(V, line_list[j], word_index))
X_train = np.asarray(X_train)
Y_train = np.asarray(Y_train)
print(X_train.shape)
print(Y_train.shape)
#print(data)
```

```python
print(V)
print(word_index)
print(Y_train.shape)
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D, AveragePooling2D, Dropout
from keras.utils import np_utils
model = Sequential()
model.add(Flatten(input_shape = (V, 1)))
model.add(Dense(100))
model.add(Dense(V, activation = "softmax"))
model.compile(loss = "categorical_crossentropy", 
              optimizer = "adam", 
              metrics = ["mae"])
model.fit(X_train.reshape(X_train.shape[0], X_train.shape[1], 1), Y_train.reshape(Y_train.shape[0],  Y_train.shape[1], 1), batch_size = 32, epochs = 30, verbose = 1, shuffle = True)
```

```python
word = input()
print(word)
top_number = 5
test = to_numeric(V, word, word_index)
test = test.reshape(1, -1, 1)
output = model.predict(test)[0]
top_ten = np.argpartition(output, -top_number)[-top_number:]
for index in top_ten:
  print("{} : {}".format(index_word[index], output[index]))
```
